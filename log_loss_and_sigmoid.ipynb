```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Log loss for a single prediction
def log_loss_single(y, y_pred):
    epsilon = 1e-15  # avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.log(y_pred) if y == 1 else -np.log(1 - y_pred)

# Generate predictions from 0.001 to 0.999
y_preds = np.linspace(0.001, 0.999, 1000)

# Compute log losses for y=1 and y=0
log_loss_y1 = [-np.log(y_pred) for y_pred in y_preds]
log_loss_y0 = [-np.log(1 - y_pred) for y_pred in y_preds]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(y_preds, log_loss_y1, label='Log Loss (y=1)', color='blue')
plt.plot(y_preds, log_loss_y0, label='Log Loss (y=0)', color='red')
plt.title('Binary Log Loss vs Predicted Probability')
plt.xlabel('Predicted Probability (ŷ)')
plt.ylabel('Log Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot Sigmoid Function
# ----------------------
z_vals = np.linspace(-10, 10, 1000)
sigmoid_vals = sigmoid(z_vals)

plt.subplot(1, 2, 2)
plt.plot(z_vals, sigmoid_vals, label='Sigmoid(z)', color='green')
plt.title('Sigmoid Function')
plt.xlabel('z')
plt.ylabel('sigmoid(z)')
plt.grid(True)
plt.axhline(0.5, linestyle='--', color='gray', alpha=0.5)
plt.axvline(0, linestyle='--', color='gray', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
